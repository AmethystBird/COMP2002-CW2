```python
#General
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

#Miscellaneous
#import random
import numpy as np
import sys

#Regression
    #General
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

    #Regressors
from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.svm import SVR

#CODE vv
#Data importing
print("Task 1.1 | Data Preparation")
#skipfooter=523 if blank rows are an issue
inputData = pd.read_excel("ENB2012_data.xlsx", usecols="A,B,C,D,E,F,G,H")
targetData = pd.read_excel("ENB2012_data.xlsx", usecols="I,J")
SVRTargetData = pd.read_excel("ENB2012_data.xlsx", usecols="I")
#print(inputData)
#print(targetData)

#Conversion to array
targetDataArray = np.array(targetData)
SVRTargetDataArray = np.array(SVRTargetData)
SVRTargetDataTrueArray = np.ravel(SVRTargetDataArray)
#print(SVRTargetDataTrueArray)

#Normalisation
normaliser = MinMaxScaler()
normalisedInputData = normaliser.fit_transform(inputData)
#print(normalisedInputData)

#Random Forest Regressor
print("Random Forest Regressor:")
RFRegressor = RandomForestRegressor()
RFRegressor.fit(normalisedInputData, targetDataArray)
RFROutput = RFRegressor.predict(normalisedInputData)
print(mean_absolute_error(targetDataArray, RFROutput))

#RFRegressor = RandomForestRegressor(max_depth=2, random_state=0)

#MLPRegressor
print("MLPRegressor:")
regressor = MLPRegressor(max_iter=100000)
regressor.fit(normalisedInputData, targetDataArray)
MLPOutput = regressor.predict(normalisedInputData)
print(mean_absolute_error(targetDataArray, MLPOutput))

#SVR Regressor
print("SVR Regressor:")
SVRRegressor = make_pipeline(StandardScaler(), SVR(C=1.0, epsilon=0.2))
SVRRegressor.fit(normalisedInputData, SVRTargetDataTrueArray)
SVROutput = SVRRegressor.predict(normalisedInputData)
print(mean_absolute_error(SVRTargetDataTrueArray, SVROutput))

#print("Custom:")
#regressor = MLPRegressor(random_state=1, max_iter=2000).fit(normalisedInputData, targetDataArray)
#outputC1 = regressor.predict(normalisedInputData)
#outputC2 = regressor.score(normalisedInputData, targetDataArray)
#print(outputC1)
#print(outputC2)

#From webbook
#regressor = MLPRegressor()
#regressor.fit(scaled, targets)
#outputs = regressor.predict(scaled)
#print(mean_absolute_error(targets, outputs))

#print("Arbitrary:")
#x, y = make_regression(n_samples=200, random_state=1)
#x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=1)
#regressor = MLPRegressor(random_state=1, max_iter=4000).fit(x_train, y_train)
#output1 = regressor.predict(x_test[:2])
#output2 = regressor.score(x_test, y_test)
#print(output1)
#print(output2)
```
